```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Importing first dataset

```python
df=pd.read_csv("combined_data_1.txt",header=None,usecols=[0,1],names=["CustomerID","Rating"])
df.head()
```

```python
df.tail()
```

### Data Wrangling

```python
# calculating movie count
movie_count = df.isnull().sum()['Rating']
movie_count
```

```python
#calculating customer count
customer_count= df["CustomerID"].nunique() -movie_count
customer_count
```

```python
rating_count=len(df)-movie_count
rating_count
```

```python
#appending movie ID
movie_id=None
movie_np=[]
for cust_id in df["CustomerID"]:
  if ':' in cust_id:
    movie_id=int(cust_id.replace(":",""))
  movie_np.append(movie_id)
```

```python
movie_np
```

```python
df["Movie_ID"]=movie_np
df
```

```python
#Cleaning the datset
df.dropna(inplace=True)
df
```

- Now we have a cleaned dataset with customer id, rating and movie count

### Preparing the data for Modelling

```python
df["CustomerID"]=df["CustomerID"].astype(int)
```

```python
#Exluding the movies with less rating count
filter_movies = df.groupby("Movie_ID")["Rating"].count()
filter_movies
```

```python
included_movies = round(filter_movies.quantile(0.6))
print(included_movies)
```

```python
Rejected_movies = filter_movies[filter_movies < included_movies].index
Rejected_movies
```

```python
# Exclusing customers who who rated less movies
filter_customers = df.groupby("CustomerID")["Rating"].count()
filter_customers
```

```python
included_customers = round(filter_customers.quantile(0.6))
included_customers
```

```python
Rejected_customers=filter_customers[filter_customers<included_customers].index
Rejected_customers
```

```python
df = df[~df["Movie_ID"].isin(Rejected_movies)]
df=df[~df["CustomerID"].isin(Rejected_customers)]
df
```

### Building Recomendation engine

```python
reader=Reader()
data_1 = Dataset.load_from_df(df[["CustomerID","Movie_ID","Rating"]][:1000000],reader)
```

```python
model=SVD()
```

```python
cross_validate(model,data_1,measures=["RMSE"],cv=3)
```
