## Top Nutritional foods

```python
# Top 10 foods by protein content
top_protein_foods = df[['Food', 'Protein (g)']].sort_values(by='Protein (g)', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='Food', y= 'Protein (g)', data= top_protein_foods, palette='viridis')
plt.title('Top 10 Foods by Protein Content')
plt.xlabel('Protein (g)')
plt.ylabel('Food')
plt.show()
```

```python
# Top 10 low calory foods
top_low_calory_foods = df[['Food', 'Calories']].sort_values(by='Calories', ascending=True).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='Food', y= 'Calories', data= top_low_calory_foods, palette='Pastel1')
plt.title('top low calory foods')
plt.xlabel('Calories')
plt.xticks(rotation=90)
plt.ylabel('Food')
plt.show()
```

```python
# Top 10 foods by fiber content
high_fiber_foods = df[['Food', 'Fiber (g)']].sort_values(by='Fiber (g)', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Fiber (g)', y='Food', data=high_fiber_foods, palette='plasma')
plt.title('Top 10 High Fiber Foods')
plt.xlabel('Fiber (g)')
plt.ylabel('Food')
plt.show()
```

```python
# Top 10 foods by antioxidant score
top_antioxidant_foods = df[['Food', 'Antioxidant Score']].sort_values(by='Antioxidant Score', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Antioxidant Score', y='Food', data=top_antioxidant_foods, palette='coolwarm')
plt.title('Top 10 Foods by Antioxidant Score')
plt.xlabel('Antioxidant Score')
plt.ylabel('Food')
plt.show()
```

```python
# Create a protein-to-calorie ratio
df['Protein/Calorie Ratio'] = df['Protein (g)'] / df['Calories']

# Top 10 foods with high protein and low calories
high_protein_low_calorie_foods = df[['Food', 'Calories', 'Protein (g)', 'Protein/Calorie Ratio']].sort_values(by='Protein/Calorie Ratio', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Protein/Calorie Ratio', y='Food', data=high_protein_low_calorie_foods, palette='cubehelix')
plt.title('Top 10 High Protein, Low-Calorie Foods')
plt.xlabel('Protein/Calorie Ratio')
plt.ylabel('Food')
plt.show()
```

## Nutritional comparison with regions

```python
# Group the data by 'Originated From' and calculate the mean of nutritional values
nutrition_by_origin = df.groupby('Originated From').mean()[['Calories', 'Protein (g)', 'Fiber (g)', 'Vitamin C (mg)', 'Antioxidant Score']]

# Reset index for easier plotting
nutrition_by_origin = nutrition_by_origin.reset_index()

# Sort the values by 'Calories' in descending order
nutrition_by_origin = nutrition_by_origin.sort_values(by='Calories', ascending=False)

# Plot the comparison for different origins (sorted in descending order)
plt.figure(figsize=(12, 6))
sns.barplot(x='Originated From', y='Calories', data=nutrition_by_origin, palette='Set2')
plt.xticks(rotation=90)
plt.title('Nutritional Comparison by Origin (Calories)')
plt.xlabel('Originated From')
plt.ylabel('Average Calories (per 100g)')
plt.show()
```

```python
protien_by_origin = nutrition_by_origin.sort_values(by='Protein (g)', ascending=False)

# Plot the comparison for different origins (sorted in descending order)
plt.figure(figsize=(12, 6))
sns.barplot(x='Originated From', y='Protein (g)', data=protien_by_origin, palette='Set2')
plt.xticks(rotation=90)
plt.title('Nutritional Comparison by Origin (Protein (g))')
plt.xlabel('Originated From')
plt.ylabel('Average Calories (per 100g)')
plt.show()
```

```python
# Sort the values by 'Fiber (g)' in descending order
nutrition_by_origin = nutrition_by_origin.sort_values(by='Fiber (g)', ascending=False)

# Plot the comparison for different origins (sorted in descending order)
plt.figure(figsize=(12, 6))
sns.barplot(x='Originated From', y='Fiber (g)', data=nutrition_by_origin, palette='Set2')
plt.xticks(rotation=90)
plt.title('Nutritional Comparison by Origin (Fiber)')
plt.xlabel('Originated From')
plt.ylabel('Average Fiber (g) per 100g')
plt.show()
```

```python
nutrition_by_origin = nutrition_by_origin.sort_values(by='Vitamin C (mg)', ascending=False)

# Plot the comparison for different origins (sorted in descending order)
plt.figure(figsize=(12, 6))
sns.barplot(x='Originated From', y='Vitamin C (mg)', data=nutrition_by_origin, palette='Set2')
plt.xticks(rotation=90)
plt.title('Nutritional Comparison by Origin (Vitamin C)')
plt.xlabel('Originated From')
plt.ylabel('Average Vitamin C (mg) per 100g')
plt.show()
```
